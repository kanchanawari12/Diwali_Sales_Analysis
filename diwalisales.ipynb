# EXPLORATORY DATA ANALYSIS (EDA)

```python
df.columns
```

### Gender

```python
ax=sns.countplot(x="Gender",data=df,palette=["blue","orange"],hue="Gender")     # count the frequency of gender and automaticallty creates barchart
for i in ax.containers:                  # ax.containers contains bar object
    ax.bar_label(i)                     #adds the count value on top of each bar
```

#### from above graph we can see that most of the buyers are female 

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,palette=["blue","orange"],hue="Gender")
```

#### above graph shows total sales by gender

## Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

#### from aabove graph we can see that most of the buyer if of age between  26-35

### state

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively


### Marital status


```python
ax = sns.countplot(data = df, x = 'Shadi',palette=["blue","orange"],hue="Shadi")

plt.figure(figsize=(7,5))   #sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Shadi', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Shadi',y= 'Amount', hue='Gender')
```

### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

## Occupation

```python
colors = [ "red","blue","green","orange","purple","brown","pink","gray","olive","cyan","#FF5733","#33FF57","#3357FF","#F1C40F","#9B59B6"]
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',palette=colors,hue="Occupation")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
colors = [ "red","blue","green","orange","purple","brown","pink","gray","olive","cyan","#FF5733","#33FF57","#3357FF","#F1C40F","#9B59B6"]
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',palette=colors,hue="Occupation")
```

### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
colors = ["red","blue","green",
    "#4C72B0", "#55A868", "#C44E52", "#8172B3", "#CCB974",
    "#64B5CD", "#8C8C8C", "#E17C05", "#937860", "#DA8BC3",
    "#8C564B", "#BCBD22", "#17BECF", "#AEC7E8", "#98DF8A"
]
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category',palette=colors,hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
colors=colors = ["red","blue","green","#4C72B0", "#55A868", "#C44E52", "#8172B3", "#CCB974", "#64B5CD", "#8C8C8C"]
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',palette=colors,hue="Product_Category")
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders',palette=colors,hue='Product_ID')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
